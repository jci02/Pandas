````python
import pandas as pd
reviews = pd.read_csv("winemag-data-130k-v2.csv",index_col=0)
````

````python
print(f"Data Type of column \"price\" is {reviews["price"].dtype}\n")
print(f"Data Types of every columns are\n{reviews.dtypes}\n")
print(f"Convert datatype of column \"points\"\n{reviews["points"].astype("float64")}\n")
print(f"Convert datatypes of columns \"ponts\" and \"price\":\n")
reviews.astype({"points":"complex", "price":"bool"})
````

````python
print(f"Index object of reviews dataframe\n{reviews.index}")
print(f"Data Type of the Index object of reviews dataframe:{reviews.index.dtype}")
print(f"Subset of dataframe with datatype in \"include\":\n{reviews.select_dtypes(include="number")}")
````

# <u>Missing Values</u>

| Action         | Code                        |
| -------------- | --------------------------- |
| detect missing | `df.isna()`                 |
| detect non missing | `df.notna()`           |
| detect missing | `df.isnull()`               |
| detect noin missing |  `df.notnull()`        |
| fill missing with value  | `df.fillna(value,method='ffill',axis)`|
| drop column           | `df.drop("col", axis=1, inplace = False)`     |
| drop column and update df           | `df.drop("col", axis=1, inplace = True)`     |
| drop multiple columns | `df.drop(subset=["A","B"], axis=1, inplace = False)` |
| drop row              | `df.drop(3)`                 |
| drop NA rows          | `df.dropna()`                |
| drop rows with only Nas          | `df.dropna(how="all")`                |
| drop NA columns       | `df.dropna(axis=1)`          |
| drop NA rows       | `df.dropna(axis=0)`          |

````python
df = pd.read_csv("Country.csv")
print(f"{df}\n")
print("Drop column \"Purchased\" but leave df unchanged: \n",df.drop("Purchased", axis=1))
````

````python
print("Drop columns \"Purchased\" and \"Age\" but leave df unchanged: \n",df.drop(["Purchased", "Age"], axis=1),"\n")
print("Drop fourth row but leave df unchanged: \n",df.drop(4, axis=0))
````

````python
print("Drop first and fourth row but leave df unchanged: \n",df.drop([1,4]),"\n")
print(f"Drop na rows: {df.dropna()}")
````

````python
print(f"Drop only complete na rows: {df.dropna(how="all")}")
print(df,"\n")
print(f"Drop na columns (\"Age\" and \"Salary\"): \n{df.dropna(axis=1)}")
````

````python
print("Drop columns \"Purchased\" and \"Age\" and update df: \n",df.drop(["Purchased", "Age"], axis=1, inplace=True))
print(df)
````

````python
print(f"Detect missing values: \n{df.isna()}\n")
print(f"Fill missing values with 0: \n{df.fillna(0)}\n")
df
````

````python
df = pd.read_csv("Country.csv")
print(df,"\n")
print(f"Replace NaNs in multiple columns:\n{df.fillna({'Age': df['Age'].mean(),'Salary': 'Unknown'})}")
````
