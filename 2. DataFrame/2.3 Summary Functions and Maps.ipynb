# <u>SummaryFunctions</u>


```python
import pandas as pd
df = pd.DataFrame(
    data =  {
        'key0': [a,b,c,...,z], # column key0
        'key1': [a,b,c,...,z], # column key1
        ...
        'keyn': [a,b,c,...,z] # column keyn
            },
    index, # len(index) == len([a,b,c,...,z])
    dtype="..."
) 

df.info() # shows column types + memory
df.describe(include="...") # shows statistics (mean, std, etc.) and if include="all" then categorical columns will be included


```
| Function      | What it computes                    | Default behavior     | Important notes                                          |
| ------------- | ----------------------------------- | -------------------- | -------------------------------------------------------- |
| `df.mean(numeric_only,axis)`   | Arithmetic mean of values           | Column-wise (axis=0) | Ignores NaN by default; works on numeric columns only    |
| `df.sum(numeric_only,axis)`    | Sum of values                       | Column-wise (axis=0) | Concatenates strings if column is object; skips NaN      |
| `df.min(numeric_only,axis)`    | Minimum value                       | Column-wise (axis=0) | Works for numeric, datetime, and strings (lexicographic) |
| `df.max(numeric_only,axis)`    | Maximum value                       | Column-wise (axis=0) | Works for numeric, datetime, and strings                 |
| `df.median(numeric_only,axis)` | Median (50th percentile)            | Column-wise (axis=0) | Numeric columns only; ignores NaN                        |
| `df.std(numeric_only,axis)`    | Standard deviation (sample, ddof=1) | Column-wise (axis=0) | Measures spread; ignores NaN                             |
| `df.var(numeric_only,axis)`    | Variance (sample, ddof=1)           | Column-wise (axis=0) | Square of std; ignores NaN                               |
| `df.count(numeric_only,axis)`  | Number of non-NaN values            | Column-wise (axis=0) | Counts valid entries per column                          |


````python
import pandas as pd
reviews = pd.read_csv("winemag-data-130k-v2.csv")
#help(reviews.mean)
````

````python
print(f"Info on dataframe:\n")
reviews.info() # only columns
````

````python
print(f"Statistics on dataframe only on numerical columns \"points\" and \"price\":\n{reviews.describe()}\n")
print(f"Statistics on dataframe only on numerical column \"points\":\n{reviews["points"].describe()}\n")
print(f"Statistics on dataframe only on categorical column \"taster_name\":\n{reviews["taster_name"].describe()}\n")
````

````python
print(f"Mean of column \"points\":{reviews.points.mean()}\n")
print(f"Unique values in column \"taster_name\":\n{reviews.taster_name.unique()}\n")
print(f"Unique values counts in column \"taster_name\":\n{reviews.taster_name.value_counts()}\n")
````

````python
import numpy as np
df = pd.DataFrame({
    "Age": [25, 30, np.nan, 40],
    "Salary": [50000, 60000, 55000, np.nan],
    "Score": [80, 90, 85, 88],
    "Name": ["Alice", "Bob", "Charlie", "David"]
})

print(df,"\n")
print(f"Column wise mean:\n{df.mean(numeric_only=True)}\n")
print(f"Row wise mean:\n{df.mean(numeric_only=True,axis=1)}\n")
print(f"Column wise sum:\n{df.sum(numeric_only=True)}\n and row wise sum:\n{df.sum(numeric_only=True,axis=1)}\n")
````

````python
print(f"Column wise minimum:\n{df.min(numeric_only=True)}\n and row wise minimum:\n{df.min(numeric_only=True,axis=1)}\n")
print(f"Column wise maximum:\n{df.max(numeric_only=True)}\n and row wise maximum:\n{df.max(numeric_only=True,axis=1)}\n")
print(f"Column wise median:\n{df.median(numeric_only=True)}\n and row wise median:\n{df.median(numeric_only=True,axis=1)}\n")
````

````python
print(f"Column wise standard deviation:\n{df.std(numeric_only=True)}\n and row wise standard deviation:\n{df.std(numeric_only=True,axis=1)}\n")
print(f"Column wise variance:\n{df.var(numeric_only=True)}\n and row wise variance:\n{df.var(numeric_only=True,axis=1)}\n")
print(f"Column wise count:\n{df.count()}\n and row wise count:\n{df.count(axis=1)}\n")
````

# <u>Maps</u>

**map() $-$ Series only and best for element-wise replacement**

**apply() $-$ flexible and works on Series (element-wise) and DataFrame (row/column-wise)**

```python
df['col'].map(func) # Returns new Series where elements from column "col" are mapped onto new values


df['age'].apply(func,axis=0) # returns new Series (same as df['age'].map(func) but general)

df.apply(func,axis="columns") # returns new Dataframe with changes according to func

# Vectorised Operations (preferred over maps)
df['age'] + 1
df['salary'] * 1.1


```

````python
points_mean=reviews.points.mean()
print(f"Remean column \"points\":\n{reviews["points"].map(lambda p: p - points_mean)}\n")

# Faster way of remeaning
points_mean=reviews.points.mean()
print("Faster remeaning\n",reviews.points - points_mean,"\n")

print(f"Rename \"US\" and \"Italy\":\n{reviews.country.map({"US":"United States", "Italy":"Germany"})}\n")
````

````python
points_mean=reviews.points.mean()
def remean_points(row):
    row["points"] = row["points"] - points_mean 
    return row

reviews.apply(remean_points,axis="columns").head()
````

````python
print(f"Concatenate countries and there regions:\n{reviews["country"] + " - " + reviews["region_1"]}\n")
bargain_idx = (reviews.points / reviews.price).idxmax()
print(f"Title of wine with highest points-to-price ration:{reviews.loc[bargain_idx,"title"]}\n")
````

````python
n_tropical = reviews["description"].map(lambda descr: "tropical" in descr).sum()
n_fruity = reviews["description"].map(lambda descr: "fruity" in descr).sum()
print(f"Counts of appearances of words \"tropical\" or \"fruity\"\n{pd.Series([n_tropical,n_fruity], index=["Tropical","Fruity"])}")
````
