# <u>Grouping</u>

```python
# Split data by single column
df.groupby(by="city") # returns GroupBy object but does nothing yet

# Split data by multiple columns
df.groupby(by=['department', 'gender']) # Creates hierarchical groups

# Group by index
df.groupby(level=0)

# Single aggregation after grouping
df.groupby('department')['salary'].mean()

# Multiple aggregations after grouping
df.groupby('department')['salary'].agg(['mean', 'max', 'min'])

# Multiindex
df.groupby(by=['department', 'gender']).description.agg([len])

# Different agg per column
df.groupby("city").agg({
    "age": ["mean", "max"],
    "salary": "sum"
})

# Named aggregations
df.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    max_age=('age', 'max')
)

# Group filtering
df.groupby('dept').filter(lambda g: len(g) > 5)

# Common groupby operations
df.groupby('dept').size()
df.groupby('dept').count()
df.groupby('dept')['col'].nunique()

```

````python
import pandas as pd
reviews = pd.read_csv("winemag-data-130k-v2.csv",index_col=0)
````

````python
reviews.groupby(by="points")["points"] # does nothing yet
````

````python
print(f"Count occurences of points:\n{reviews.groupby(by="points")["points"].count()}\n")
print(f"Count occurences of points:\n{reviews["points"].value_counts()}\n")
````

````python
print(f"Get cheapest wine in each point value category:\n{reviews.groupby("points").price.min()}")
````

````python
print(f"Select name of the first wine reviewed from each winery:\n{reviews.groupby("winery").apply(lambda df: df.title.iloc[0],include_groups=False)}")
````

````python
print(f"Pick best wine by country and province:\n")
reviews.groupby(["country","province"]).apply(lambda df:df.loc[df.points.idxmax()],include_groups=False)
````

````python
print("Under column len are the numbers of price occurences for each country like Argentina:",len(reviews.loc[reviews.country=="Argentina","price"]))
print(f"Get {reviews.groupby(["country"]).price.agg([len,"min","max"])}")
````

````python
df = pd.read_csv("Country.csv")
print(df,"\n")
print(f"Calculate mean of \"Age\" for the only countries France, Germany and Spain: \n{df.groupby("Country")["Age"].mean()}\n")
print(f"Count occurances for the only countries France, Germany and Spain: \n{df.groupby("Country")["Age"].count()}\n")
print(f"Calculate mean and minimum of \"Age\" and sum of salary for every country: \n{df.groupby("Country").agg({"Age":["mean", "min"],"Salary":["sum"]})}\n")
print(f"Calculate mean and minimum of \"Age\" and sum of salary for every country: \n{df.groupby("Country").agg(Age_stats=("Age", "mean"),Salary_sum=("Salary","sum"))}\n")
````

````python
print(f"Multiindex:\n")
reviews.groupby(["country","province"]).description.agg([len])
````

# <u>Sorting</u>

```python
# Sort rows by a column
df.sort_values(by="age",axis,na_position,key)

# Descending
df.sort_values(by="age", ascending=False)

# Sort by multiple columns
df.sort_values(["city", "age"])

# Different order per column
df.sort_values(['dept', 'salary'], ascending=[True, False])

# Handle NaNs
df.sort_values('salary', na_position='first')

# In-place sort (usually avoid since chaining becomes harder)
df.sort_values('salary', inplace=True)

# Sort by row labels
df.sort_index()

# Sort columns by name
df.sort_index(axis=1)

# MultiIndex sorting
df.sort_index(level=0)

```

````python
df = pd.read_csv("Country.csv")
print(df, "\n")
print(f"Sorted by column \"Age\" and NaN are put last:\n{df.sort_values(by="Age")}\n")
````

````python
print(f"Sorted by column \"Age\" reverse and NaN are put last:\n{df.sort_values("Age", ascending=False)}\n")
print(f"Sorted by columns \"Salary\" (first) and \"Age\" and NaN are put last:\n{df.sort_values(["Salary", "Age"])}\n")
````
