# Core Object of Pandas

DataFrame $\rightarrow$ 2-dimensional labeled table (collection of Series sharing the same index and having the column name as name of the Series)

| Series        | DataFrame              |
| ------------- | ---------------------- |
| 1-D           | 2-D                    |
| One column    | Multiple columns       |
| Has one index | Has row + column index |
| Used in Data Frames       | Columns of DataFrames are a Series

**A DataFrame is essentially a dict-like container of aligned Series. Each column in a DataFrame is a Series.**

### Inspecting and Exploring Dataframes

| Useful Functions        | What it does                 |
| --------------- | ---------------------------- |
| `df.head()`     | shows first 5 rows                 |
| `df.tail()`     | shows last 5 rows                  |
| `df.info()`     | shows column types + memory        |
| `df.describe()` | shows statistics (mean, std, etc.) |
| `df.shape`      | (num_rows, num_columns)              |
| `df.columns`    | displays column names                 |
| `df.dtypes`     | displays data types per column                   |
| `df.copy()`     | shallow copies df |
|  **Indexmanipulation**  | **What it does** |
| `df.index`      | displays row index labels                   |
| `df.index.name`  | stores the name of the index axis  |
| `df.rename_axis("...",axis="columns",inplace=True)`| Renames the index or columns axis label    |
| `df.reindex(index,labels,columns)`  | Conforms the DataFrame to a new set of index (or columns) labels. Can add, remove, or reorder rows/columns   |
| `df.set_index("id",in_place=True)` | Replaces index labels of df with values of column "id" |
| `df.reset_index(inplace=True)` | moves non-default index object into normal column and restores default index | 

````python
df5 = pd.DataFrame(
    zip(["Jack", "Dan"],[23,25],["London","Manchester"]),
    columns=["name","age","city"],
    index=["Person1", "Person2"])
print(df5,"\n")
print("Description of df5:\n",df5.describe(),"\n")
print("Columns of df5:",df5.columns,"\n")
print("Index labels:",df5.index,"\n")
print("Data type per column:\n",df5.dtypes,"\n")
print("Set column \"name\" as index but leave original unchanged:\n",df5.set_index("name"),"\n")
df5.set_index("name",inplace=True) # Set column "name" as index and update original
print("New original DataFrame df5:\n",df5,"\n")
print("Restore column \"name\" and set default index but leave original unchanged:\n",df5.reset_index(),"\n")
df5.reset_index(inplace=True) # Restore column \"name\" and set default index and update original
print("New original DataFrame df5:\n",df5,"\n")
````

````python
df5 = pd.DataFrame(
    zip(["Jack", "Dan"],[23,25],["London","Manchester"]),
    columns=["name","age","city"],
    index=["Person1", "Person2"])
print(df5,"\n")
print("Index name of", df5.index,"of df5 is:",df5.index.name)
df5.index.name = "Personas" # change original to have index name "Personas"
print("Index name of", df5.index,"of df5 is:",df5.index.name)
print(df5,"\n")
print("Change index name not in place (original unchanged):\n",df5.rename_axis("row_id",axis="rows")) # same affect as df5.index.name = "row_id"
df5.rename_axis("row_id",inplace=True) # change original to have index name "row_id"
print(df5,"\n")
print("Column name of", df5.columns,"of df5 is:",df5.columns.name)
df5.rename_axis("features", axis="columns",inplace=True) # change column name in place
print("New original:\n",df5,"\n")
print("Column name of", df5.columns,"of df5 is:",df5.columns.name,"\n")
print("Reorder rows\n",df5.reindex(index=["Person2","Person1"]),"\n")
print("Reorder rows and add new row\n",df5.reindex(["Person2","Person1","Person3"]),"\n")
print("Remove a row\n",df5.reindex(["Person2"]),"\n")
print("Reorder columns, remove a column and add one\n",df5.reindex(columns=["city","age","job"]),"\n")
````

````python
df6 = pd.DataFrame(
    {
        "Name":["Alice","Bob","Charlie"],
        "Age":[25,30,35],
        "Salary":(50000,60000,70000),
        "City":["Berlin","Paris","New York"],
    }
)

df6_shallow_copy = df6.copy() # Changing df6_shallow_copy does NOT change df6 and vice versa
print("Both Dataframes are the same:",all(df6_shallow_copy == df6))
print("Both variables point to same DataFrame in memory:",df6_shallow_copy is df6,"\n")
df6_shallow_copy["Age"] = "Twentyfive"
print(df6, "\n")
print(df6_shallow_copy, "\n")


df6_deep_copy = df6 # both variables point to the same DataFrame in memory and any change to df6_deep_copy will change df6 too
print("Both Dataframes are the same:",all(df6_deep_copy == df6))
print("Both variables point to same DataFrame in memory:",df6_deep_copy is df6,"\n")
df6_deep_copy["Age"] = "Twentythree"
print(df6, "\n")
print(df6_deep_copy, "\n")
````
