# <u>Selecting</u>

```python
import pandas as pd
df = pd.DataFrame(data,index,columns)
```
### COLUMN SELECTION

```python
# Single column by column index
df[3] # -> returns a Series

# Single column by name
df['col'] # -> returns a Series
df.col # -> returns a Series and only if column name has no spaces/special charatcers

# Multiple columns by name list
df[['col1', 'col2',...,'colk']] # -> returns a DataFrame

```
**``.loc`` is label-based selection and when used with slicing the first element of the range is <u>included</u> as well as the last**

```python

# Multiple columns by name list
df.loc[:, ['col1', 'col2',...,'colk']] # -> returns a DataFrame

# Multiple columns by name (with Slicing -> col4 included)
df.loc[:, 'col1':'col4':step] # -> returns a DataFrame
```
**``.iloc`` is position-based selection and when used with slicing the first element of the range is <u>included</u> and the last <u>excluded</u>**

### ROW SELECTION

```python

# Single row by label
df.loc[label] # -> returns a Series

# Single row by label
df.loc[[label]] # -> returns a DataFrame

# Single row by integer position
df.iloc[3] # -> returns a Series

# n random rows
df.sample(n)

# Multiple rows by position
df[start:stop:step] # -> returns a DataFrame

# Multiple rows by label list
df.loc[[label1, label2]] # -> returns a DataFrame

# Multiple rows by position list
df.iloc[[0, 3, 5]] # -> returns a DataFrame

# Multiple rows by label slicing
df.loc[start:end:step] # -> returns a DataFrame

# Multiple rows by position slicing
df.iloc[2:5:1] # -> returns a DataFrame

```
**Both ``.loc`` and ``.iloc`` are row-first, column-second: ``.loc[row,column]`` and ``.iloc[row,column]``**

### COLUMN + ROW SELECTION
```python
# Single row + single column (label based)
df.loc[row_label, 'col'] # -> single entrie of dataframe

# Single row + single column (position based)
df.iloc[0, 4] # -> single entrie of dataframe

# Mulitple rows + multiple columns (label based)
df.loc[label1:label5, ['col1', 'col2', 'col3']]

# Mulitple rows + multiple columns (position based)
df.loc[1:5, 1:3]

# Fast single-value access
df.at[row_label2, col4] # label based
df.iat[2, 4] # position based

```

# <u>Filtering</u> (Boolean Indexing)

``&`` (and)

``|`` (or)

``~`` (not)

```python
import pandas as pd
df = pd.DataFrame(
    data =  {
        'key0': [a,b,c,...,z], # column key0
        'key1': [a,b,c,...,z], # column key1
        ...
        'keyn': [a,b,c,...,z] # column keyn
            },
    index, # len(index) == len([a,b,c,...,z])
    dtype="..."
)

# Select rows based on a condition
df[df["age"] > 20]
df[df["name"] == "Alice"]
df[(df["A"] > 0) & (df["B"] < 10)]


# Use .isin() to select rows with specific value inside a column
df[df['country'].isin(['USA', 'Canada'])]

# Select only rows fullfilling a specific condition inside a column
df[df["points"].between(85,90)]

# Return the elements in the given indices along an axis
df.take(indices,axis)

# Query Syntax: Human-readable selection using strings
df.query("age > 20")
df.query("city == 'Berlin'")
df.query("A > 5 and B < 10")
df.query('age > 30 and score > 80')


# Using .filter()
df.filter(like="name") # Columns containing substring
df.filter(regex="^A_") # Columns matching regex
# Selecting rows or columns by axis
df.filter(items=["col1", "col2"])
df.filter(items=["row3", "row5"], axis=0)


# Return the first n rows with the smallest values in columns in
# ascending order
df.nsmallest(n,columns)

# Return the first n rows with the largest values in columns in
# descending order
df.nlargest(n,columns)


```

````python
print(f"Appearences of country \"Italy\" in column \"country\":\n {reviews.country=="Italy"}\n")
print(f"Select only rows with \"Italy\" as country:\n")
reviews.loc[reviews.country=="Italy"]
````

````python
print(f"Select rows with \"Italy\" as country and points greater than 90:\n")
reviews[(reviews.country=="Italy") & (reviews.points >= 90) ]
````

````python
print(f"Select rows with \"Italy\" as country or points greater than 90:\n")
reviews[(reviews["country"]=="Italy") | (reviews["points"] >= 90) ]
````

````python
print(f"Select only rows with \"Italy\" or \"France\" as country:")
reviews[reviews["country"].isin(["Italy","France"])]
````

````python
print(f"Select rows with points greater than 30 and only look at prices:\n{reviews.loc[reviews['points'] > 30, 'price']}\n")
print(f"Select only rows with points between 85 and 90:")
reviews[reviews["points"].between(85,90)]
````

````python
print("Take rows 10 and 1:")
reviews.take([10,1],axis=0)
````

````python
print("Take last row and the one before last:")
reviews.take([-1,-2])
````

````python
print("Take columns 10 and 1:")
reviews.take([10,1],axis=1)
````

````python
print("5 largest values in row \"price\" in descending order:")
reviews.nlargest(5, 'price')
````

````python
print("15 smallest values in row \"price\" in ascending order:")
reviews.nsmallest(15, 'price')
````
