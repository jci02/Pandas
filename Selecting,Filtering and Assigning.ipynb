# <u>Selecting</u>

```python
import pandas as pd
df = pd.DataFrame(data,index,columns)
```
### COLUMN SELECTION

```python
# Single column by name
df['col'] # -> returns Series
df.col # -> returns Series and only if column name has no spaces/special charatcers

# Multiple columns by name list
df[['col1', 'col2',...,'colk']] # -> returns DataFrame

```
**``.loc`` is label-based selection and when used with slicing the first element of the range is <u>included</u> as well as the last**

```python

# Multiple columns by name list
df.loc[:, ['col1', 'col2',...,'colk']] # -> returns DataFrame

# Multiple columns by name (with Slicing -> col4 included)
df.loc[:, 'col1':'col4':step] # -> returns DataFrame
```
**``.iloc`` is position-based selection and when used with slicing the first element of the range is <u>included</u> and the last <u>excluded</u>**

### ROW SELECTION

```python
# Single row by label
df.loc[label] # -> returns Series

# Single row by integer position
df.iloc[3] # -> returns Series

# Multiple rows by position
df[start:stop:step] # -> returns DataFrame

# Multiple rows by label list
df.loc[[label1, label2]] # -> returns DataFrame

# Multiple rows by position list
df.iloc[[0, 3, 5]] # -> returns DataFrame

# Multiple rows by label slicing
df.loc[start:end:step] # -> returns DataFrame

# Multiple rows by position slicing
df.iloc[2:5:1] # -> returns DataFrame
```
**Both ``.loc`` and ``.iloc`` are row-first, column-second: ``.loc[row,column]`` and ``.iloc[row,column]``**

### COLUMN + ROW SELECTION
```python
# Single row + single column (label based)
df.loc[row_label, 'col'] # -> single entrie of dataframe

# Single row + single column (position based)
df.iloc[0, 4] # -> single entrie of dataframe

# Mulitple rows + multiple columns (label based)
df.loc[label1:label5, ['col1', 'col2', 'col3']]

# Mulitple rows + multiple columns (position based)
df.loc[1:5, 1:3]

# Fast single-value access
df.at[row_label2, col4] # label based
df.iat[2, 4] # position based

```

Since some of the functions return a Series and Series can be considered as the columns of a Dataframe let us quickly review Selection on a Series since it might come in handy.

```python
s = pd.Series([1,2,3], index=[a,b,c])

```
|  |  |
| ---| --- |
|a  |  1 |
|b   |  2 |
|c |    3 |



| Syntax            | Type of indexing | What it selects |
|------------------|------------------|-----------------|
| `s["a"]`           | label            | value at label `a` |
| `s[["a","b"]]`       | label list       | values at labels `a` and `b` |
| `s[3]`           | position *;often ambiguous tries labels first, but may fall back to position.This can be confusing which is why `loc` and `iloc` are preferred*| 4th value |
| `s[start:stop:step]` | slicing          | a slice by **position** |
| `s.loc["b"]`       | label-based      | value at label `b` |
| `s.loc[start:stop:step]` | label-based | slice by label |
| `s.iloc[2]`      | position-based   | 3rd value |
| `s.iloc[start:stop:step]` | position-based | slice by position |
