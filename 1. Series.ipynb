# Core Object of Pandas

Series $\rightarrow$ 1-dimensional labeled array (like a single column)


### Creating a Series

```python
pd.Series(data=None, index=None,dtype,name=None) 
# data receives list, tuple or numpy.ndarray 
# index takes a list or tuple of so called labels
# dtype stores one data type like "int", "float", "complex", "bool", "str" etc. and uses    upcasting to a common dtype

pd.Series({key1:val1, key2:val2, ...., keyn:valn}) 
# with key1,...,keyn collected in index and data=np.array([val1,val2,...,valn])

```


````python
import pandas as pd

#help(pd.Series)

import numpy as np
````

````python
# Create Series from list with labels A,B,C inside index object
s1 = pd.Series(data=[10, 20, 30],index=["A","B","C"],dtype="complex",name="s1")
print(s1,"\n")

# Create Series from numpy.ndarray with labels 1,2,3 inside index object
s2 = pd.Series(data=np.array(["Hi","there","!"]),index=[1,1,3],name="s2")
print(s2,"\n")

# Create Series from tuple with labels 10,20,30 inside index object
s3 = pd.Series(data=(3,2,1),index=[10,20,30],name="s3")
print(s3,"\n")

# Create Series from dicttionary with labels A,2,C inside index object
s4 = pd.Series({"A":10, 2:20, "C":30},dtype="float",name="s4")
print(s4,"\n")
````

````python
print(f"Deafult indexing: \n{pd.Series((10,20,3))}\n")
print(f"Custom indexing (labels) with index=[...]: \n{pd.Series([1.4,.2,3.1425,4], index=["a", "b", "c", "d"])}\n")
print(f"Custom indexing (labels) with index=(...): \n{pd.Series([1.4,.2,3.1425,4], index=("a", "b", "c", "d"))}\n")
print(f"Custom indexing from dictionary: \n{pd.Series({"First": 10-1j, 5: 20, (3+4j): 30})}\n")
print(f"Scalar value: \n{pd.Series(5, index=["first", "second", "third", "fourth"])}\n")
````

````python
# List comprehension in a Series
nested_lst = [["Hannes", "H"], ["Thomas", "T"], ["Jason", "J"]]
s = pd.Series(i[1] for i in nested_lst)
print(f"Series of nested list:\n{s}\n")
s2 = pd.Series([i[0] for i in nested_lst], index=[i[1] for i in nested_lst])
print(f"Series of nested list:\n{s2}")
````

### Index and Values of a Series

```python
import pandas as pd
s = pd.Series(data,index,dtype,name)

# Get and reassign index object
s.index

# Reorder, add or remove index labels following new_labels of type list or tuple (does not change in place!)
s.reindex(new_labels)

# Rename index labels  using a dictionary or a function (does not change in place!)
s.rename(...)

# Turns index into a column and replaces index labels with default integers; also type of s.reset_index() is pandas.core.frame.DataFrame whereas type of s is pandas.core.series.Series
s.reset_index()

# Get data object (read only)
s.values

# Get and/or reassign name of series 
s.name


# Get and/or reassign index object name
s.index.name


```

````python
s5 = pd.Series(data=[[0],[1,2,3], "Hi"],index=[1,2,3],name="s5")
print(s5,"\n")
print("Index object of s5:", s5.index,"\n")
s5.index=["A","B","C"] # reassign index object of s5
print("New Index object of s5:", s5.index,"\n")
print("New s5:\n",s5,"\n")
print("Reordered labels of s5:\n",s5.reindex(("C","A","B")),"\n")
print("Add label \"D\" with default value of NaN to s5:\n",s5.reindex(("C","A","B","D")),"\n")
print("Remove label \"A\" from s5:\n",s5.reindex(("C","B","D")),"\n")
print("------------------------------------\n")

s6 = pd.Series(data=(3,2,1),index=[1,2,3],name="s6")
print("Data object of s6:", s6.values,"\n")
print("------------------------------------\n")

s7 = pd.Series(data=([0],(1,2,3), "Hi"),index=["Martha","Nick","Holly"],name="s7")
print(s7,"\n")
print("Name object of s7:", s7.name,"\n")
s7.name="Series" # reassign name object of s7
print("New name object of s7:", s7.name,"\n")
print("Rename old labels \"Martha\" and \"Nick\" of", s7.index,"to\n",s7.rename({"Martha":"Berta", "Nick":"Mike"}),"\n")
print("Rename all old labels of", s7.index,"to\n",s7.rename(lambda x: x.lower()),"\n")
print("------------------------------------\n")

s8 = pd.Series(data=(["patient1","patient1"],"patient2", "patient3"),index=["1st","2nd","3rd"],name="s8")
print(s8,"\n")
print("Name of index object of s8:", s8.index.name,"\n")
s8.index.name = "Labels" # assign a name to index object of s8
print("New name of index object of s8:", s8.index.name,"\n")
print("Reset index", s8.index,"of s8:\n",s8.reset_index(),"\n with type=",type(s8.reset_index()))
````

### Selecting and Sclicing

```python
s = pd.Series([1,2,3], index=[a,b,c])

```
|  |  |
| ---| --- |
|a  |  1 |
|b   |  2 |
|c |    3 |



**Selecting & Slicing Table for a Series**

| Syntax            | Type of indexing | What it selects |
|------------------|------------------|-----------------|
| `s["a"]`           | label            | value at label `a` |
| `s[["a","b"]]`       | label list       | values at labels `a` and `b` |
| `s[3]`           | position *;often ambiguous tries labels first, but may fall back to position.This can be confusing which is why `loc` and `iloc` are preferred*| 4th value |
| `s[start:stop:step]` | slicing          | a slice by **position** |
| `s.loc["b"]`       | label-based      | value at label `b` |
| `s.loc[start:stop:step]` | label-based | slice by label |
| `s.iloc[2]`      | position-based   | 3rd value |
| `s.iloc[start:stop:step]` | position-based | slice by position |

````python
data = [10, 20, 30, 40, 50]
s = pd.Series(data, index=["a", "b", "c", "d", "e"])
print(f"Select value with label \"c\": {s["c"]}")
s["c"] = 2020
print(f"Modified value with label \"c\": {s["c"]}")
print(f"Select values with labels \"a\", \"c\" and \"e\": \n{s[["a", "c", "e"]]}")
#print(f"Select fourth element:{s[3]}") # use s.iloc[pos] instead since treating keys as positions is deprecated
print(f"Select fourth element:{s.iloc[3]}")
print(f"Slice from position 1 to 4 (excluded): \n{s[1:4]}")
print(f"Slice from position 1 to 4 (excluded): \n{s.iloc[1:4]}")
print(f"Select value with label \"b\": {s.loc["b"]}")
print(f"Slice from label to \"b\" to \"d\" (inclusive) in steps=2: \n{s.loc["b":"d":2]}")
````

````python
# Weakness of s[...]: integers as labels
s = pd.Series([5,3,1,2,4], index=[1,2,3,4,5])
print("s=\n{}".format(s)) 
# s[3] behaves like s.loc[3] when the index contains the label 3
print(f"The index object contains the 3 so python interprets s[3]={s[3]} as label-based indexing instead of position based ")
print(f"Therfor use s.iloc[3]={s.iloc[3]} which is correct since indexing starts from 0")
````

### Boolean Filtering

````python
s = pd.Series(list(range(1,6)), [chr(i) for i in list(range(65,70))])
print(f"s=\n{s}")
print(f"Boolean filtering (only values > 2): \n{s > 2}")
print(f"Only values > 2: \n{s[s > 2]}")
print(f"Boolean filtering (1 < values < 4): \n{s[(s > 1) & (s < 4)]}")
````

### Useful Series attributes

```python
import pandas as pd

s = pd.Series(data,index,dtype,name)

# Returns data type of values in data
s.dtype

# Returns length of data
s.size

# Returns shape of data as tuple (s.size,) since Series are 1-dimensional
s.shape

s.describe()

# Returns info on the Series such as data type, memory and size
s.info()

# Returns a boolean Series indicating NaN values in data
s.isna()

s.fillna(val)

s.dropna()
```

### Common Series Methods


| Category                  | Function                                 | Description                                                              |
| ------------------------- | ---------------------------------------- | ------------------------------------------------------------------------ |
| Math / Statistics         | `s.sum()`                                | Returns the sum of all values in the Series                              |
| Math / Statistics         | `s.mean()`                               | Returns the average (arithmetic mean)                                    |
| Math / Statistics         | `s.std()`                                | Returns the standard deviation (spread of values)                        |
| Math / Statistics         | `s.min()`                                | Returns the smallest value                                               |
| Math / Statistics         | `s.max()`                                | Returns the largest value                                                |
| Math / Statistics         | `s.median()`                             | Returns the median (50th percentile)                                     |
| Math / Statistics         | `s.var()`                                | Returns the variance (measure of dispersion)                             |
| Index operations          | `s.sort_index()`                         | Sorts the Series by index labels                                         |
| Index operations          | `s.sort_values()`                        | Sorts the Series by its values                                           |
| Value counts & uniqueness | `s.value_counts()`                       | Counts how often each unique value appears                               |
| Value counts & uniqueness | `s.unique()`                             | Returns the unique values in order of appearance                         |
| Value counts & uniqueness | `s.nunique()`                            | Returns the number of unique values                                      |
| String operations         | `s.str.lower()`                          | Converts all strings in the Series to lowercase                          |
| String operations         | `s.str.contains("abc")`                  | Checks whether each string contains `"abc"` and returns a Boolean Series |
| String operations         | `s.str.replace("a", "z")`                | Replaces occurrences of `"a"` with `"z"` in each string                  |
| Apply function            | `s.apply(lambda parameters: expression)` | Applies a custom function element-wise to each value in the Series       |


````python
import random as rand
data = [-float(rand.randint(1,10)) for i in range(0,10)]
index = [rand.randint(1,100) for i in range(0,10)]
s11 = pd.Series(data = data, index=index)
print(f"s=\n{s11}\n")
````

````python
print(f"Sum of all values is {s11.sum()}")
print(f"Mean of all values is {s11.mean()}")
print(f"Maximum of all values is {s11.max()}")
print(f"Median of all values is {s11.median()}\n")
print(f"Sort by labels:\n{s11.sort_index()}\n")
print(f"Sort by values:\n{s11.sort_values()}")
````

````python
print(f"Value counts: \n{s11.value_counts()}\n")
print(f"Unique values: {s11.unique()}\n")
print(f"Number Unique values: {s11.nunique()}\n")
print(f"Square values: \n{s11.apply(lambda x: x**2)}\n")
````

### Conversion

```python
import pandas as pd
s = pd.Series(data, index)

# Convert to NumPy array
arr = s.to_numpy() # extracts only the values of the Series and puts them in a NumPy array.

# Convert to list
lst = s.tolist() # Converts the values into a normal Python list

# Convert to dictionary
d = s.to_dict() # Converts the Series into a Python dictionary where keys = labels and values = data

```

````python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(f"s=\n{s}")
arr = s.to_numpy()
print(f"Converts to NumPy array with only values (no commas):{arr}")

lst = s.tolist()
print(f"Converts to Python list with only values (with commas):{lst}")

d = s.to_dict()
print(f"Converts to Python dictionary:{d}")

````

### Arithmetic with fill values

```python
import pandas as pd
s1 = pd.Series(data, index)
s2 = pd.Series(data, index)

s1.add(s2, fill_value=0) # Performs s1+s2 but if an index exists in one Series and not the other, pandas inserts the fill_value
s1.sub(s2, fill_value=0) # Performs s1-s2 but if an index exists in one Series and not the other, pandas inserts the fill_value

```

````python
s1 = pd.Series([1, 2, 3], index=["a", "b", "c"])
s2 = pd.Series([10, 20], index=["b", "d"])

result = s1.add(s2, fill_value=0)
print(f"Add s1 and s2 alinged by indices: \n{result}")
# a     1.0      # 1 + 0 (a is missing in s2 -> replaced with 0)
# b    22.0      # 2 + 20
# c     3.0      # 3 + 0 (no c in s2)
# d    20.0      # 0 + 20 (no d in s1)
# dtype: float64

result = s1.sub(s2, fill_value=100)
print(f"Subtract s1 and s2 alinged by indices: \n{result}")
````
