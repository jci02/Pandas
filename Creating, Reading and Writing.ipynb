# Core Objects

Series $\rightarrow$ 1-dimensional labeled array (like a single column)

DataFrame $\rightarrow$ 2-dimensional labeled table (collection of Series sharing the same index; multiple columns)

**Link: A DataFrame is essentially a dict-like container of aligned Series. Each column in a DataFrame is a Series**

# <u>Creating</u>

### Creating a Series

```python
pd.Series(data=None, index=None,dtype,name=None) 
# data receives list, tuple or numpy.ndarray 
# index takes a list or tuple
# dtype stores one data type like "int", "float", "complex", "bool", "str" etc. and uses    upcasting to a common dtype

pd.Series({key1:val1, key2:val2, ...., keyn:valn}) 
# with key1,...,keyn collected in index and data=np.array([val1,val2,...,valn])

```

````python
import pandas as pd

#help(pd.Series)

import numpy as np
````

````python
# Create Series from list
s1 = pd.Series(data=[10, 20, 30],index=[1,2,3],dtype="complex",name="s1")
print(s1,"\n")

# Create Series from numpy.ndarray
s2 = pd.Series(data=np.array(["Hi","there","!"]),index=[1,2,3],name="s2")
print(s2,"\n")

# Create Series from tuple
s3 = pd.Series(data=(3,2,1),index=[1,2,3],name="s3")
print(s3,"\n")

# Create Series from dicttionary
s4 = pd.Series({"A":10, 2:20, "C":30},dtype="float",name="s4")
print(s4,"\n")
````

### Index and Values of a Series

```python
import pandas as pd
s = pd.Series(data,index,dtype,name)

# Get and reassign index object
s.index

# Get data object (read only)
s.values

# Get and reassign name
s.name

# Get and reassign index object name
s.index.name


```

````python
s5 = pd.Series(data=[[0],[1,2,3], "Hi"],index=[1,2,3],name="s5")
print(s5,"\n")
print("Index object of s5:", s5.index,"\n")
s5.index=["A","B","C"] # reassign index object of s5
print("New Index object of s5:", s5.index,"\n")
print("------------------------------------\n")

s6 = pd.Series(data=(3,2,1),index=[1,2,3],name="s6")
print("Data object of s6:", s6.values,"\n")
print("------------------------------------\n")

s7 = pd.Series(data=([0],(1,2,3), "Hi"),index=[1,2,3],name="s7")
print(s7,"\n")
print("Name object of s7:", s7.name,"\n")
s7.name="Series" # reassign name object of s7
print("New name object of s7:", s7.name,"\n")
print("------------------------------------\n")

s8 = pd.Series(data=(["patient1","patient1"],"patient2", "patient3"),index=[1,2,3],name="s8")
print(s8,"\n")
print("Name of index object of s8:", s8.index.name,"\n")
s8.index.name = "Labels" # assign a name to index object of s8
print("New name of index object of s8:", s8.index.name,"\n")

````

### Useful Series attributes

```python
import pandas as pd

s = pd.Series(data,index,dtype,name)

# Returns data type of values in data
s.dtype

# Returns length of data
s.size

# Returns shape of data as tuple (s.size,) since Series are 1-dimensional
s.shape

# Returns a boolean Series indicating NaN values in data
s.isna()

# Opposite of s.isna() (True where no NaN values)
s.notna()

```



````python
s9 = pd.Series(data=(34,(34+0j),34.0),index=(0,1,2),name="s9")
print(s9,"\n")
print("dtype is",s9.dtype,"due to upcasting")
print("Size of s9 is",s9.size)
print("Shape of s9 is",s9.shape)
````
